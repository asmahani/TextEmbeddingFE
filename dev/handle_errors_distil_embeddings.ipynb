```python
import pandas as pd

dfEmbeddings = pd.read_csv(
    'C:/Users/alire/OneDrive/data/statman_bitbucket/aki/LLM/March2024/openai_3large_operation.csv'
)
dfEmbeddings.head()
```

```python
dfPatients = pd.read_csv(
    'C:/Users/alire/OneDrive/data/statman_bitbucket/aki/LLM/March2024/patients_for_python.csv'
)
dfPatients#.head()
```

```python
my_features = ['age', 'is_female', 'height_residual', 'bmi']

dfPatients_subset = dfPatients.loc[:, ['project_id', 'operation_no', 'kdigo_stage'] + my_features].dropna()
#dfPatients_subset#.head()
```

```python
dfCombined = pd.merge(
    dfPatients_subset
    , dfEmbeddings
    , on = ['project_id', 'operation_no']
    , how = 'inner'
)
dfCombined.head()
```

```python
X, y, Z = (
    dfCombined.iloc[:, (3 + len(my_features)):].to_numpy()
    , dfCombined.iloc[:, 2].to_numpy(dtype = 'int')
    , dfCombined.iloc[:, 3:(3 + len(my_features))].to_numpy()
)
```

```python
from sklearn.model_selection import train_test_split
import numpy as np
X_train, X_test, y_train, y_test, Z_train, Z_test = train_test_split(X, y, Z, test_size = 0.3)
```

```python
from TextEmbeddingFE.main import FeatureExtractor_BinaryClassifier

my_fe = FeatureExtractor_BinaryClassifier(n_neighbors = 50).fit(X = X_train, y = y_train, cv = 10)

Z_train_B = np.column_stack((Z_train, my_fe.predict_proba()))
Z_test_B = np.column_stack((Z_test, my_fe.predict_proba(X_test)))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import roc_auc_score

my_model_A = LogisticRegression().fit(Z_train, y_train)
my_roc_A = roc_auc_score(y_test, my_model_A.predict_proba(Z_test)[:, 1])

my_model_B = LogisticRegression().fit(Z_train_B, y_train)
my_roc_B = roc_auc_score(y_test, my_model_B.predict_proba(Z_test_B)[:, 1])
```
